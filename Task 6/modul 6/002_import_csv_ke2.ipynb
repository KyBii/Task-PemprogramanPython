```python
import pandas as pd 
import matplotlib.pyplot as plt
pd = pd.read_csv("dataSpotify.csv")
```

```python
ppb = list(pd['Pendengar per Bulan'])
nama_band = list(pd['Nama Band'])

print(nama_band)
print(ppb)
```

```python
plt.bar(nama_band, ppb)
plt.title("Jumlah Pendengar Perbulan di Spotify")
plt.xlabel("Nama Band")
plt.ylabel("Pendengar Perbulan di Spotify")
plt.show()
```
